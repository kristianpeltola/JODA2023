```python
# Tarvittavien kirjastojen lataaminen
import pandas as pd
import urllib.request
import matplotlib.pyplot as plt
import os
```

```python
# Tiedoston nimi
file_path = 'TSLA.csv'

# Tarkistetaan onko tiedosto asennettu
if os.path.isfile(file_path):
    # Ladataan tiedosto tietokoneelta
    tesla_data = pd.read_csv(file_path)
else:
    # Ladataan tiedosto internetistä
    url = 'https://query1.finance.yahoo.com/v7/finance/download/TSLA?period1=1649727142&period2=1681263142&interval=1d&events=history&includeAdjustedClose=true'
    response = urllib.request.urlopen(url)
    content = response.read()
    with open(file_path, 'wb') as f:
        f.write(content)
    tesla_data = pd.read_csv(file_path)

# Tulostetaan ensimmäiset rivit datasta
print(tesla_data.head())
```

Pythonissa on valtava määrä koneoppimisalgoritmeja, jotka on tarkoitettu monenlaisiin tehtäviin. Tässä on esimerkkejä joistakin tärkeimmistä algoritmeista ja niiden merkityksestä:

## Supervised Learning (Ohjattu oppiminen)

[Ohjattu oppiminen](https://www.javatpoint.com/supervised-machine-learning) on koneoppimisen osa-alue, jossa koneelle syötetään merkittyä dataa ja sen tehtävänä on oppia tunnistamaan syötteiden ja haluttujen vastausten välillä oleva yhteys. Tämä tarkoittaa sitä, että syötteet ja vastaukset ovat esimerkkejä, joista kone pyrkii löytämään yleisiä sääntöjä ja malleja, joita voidaan käyttää uusien vastausten ennustamiseen syötteiden perusteella.

Ohjatussa oppimisessa syötteet ja vastaukset muodostavat opetusdatan, joka jaetaan yleensä kahteen osaan: opetusdataan ja testidataan. Kone käyttää opetusdataa oppiakseen mallin, joka sitten testataan testidatalla. Testidatan avulla arvioidaan, kuinka hyvin oppimismalli toimii ennustettaessa vastauksia uusille syötteille.

### Ohjatun oppimisen algoritmeja:
- [Linear Regression](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html): Yksinkertainen algoritmi, joka etsii lineaarista suhdetta kahden muuttujan välillä ja ennustaa jatkuvia numeerisia arvoja. Sitä käytetään usein ennustamiseen ja regressio-analyysiin.
- [Logistic Regression](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html): Toimii samalla tavalla kuin lineaarinen regressio, mutta sen tavoitteena on ennustaa kahden vaihtoehdon binäärinen tulos. Sitä käytetään usein luokitteluongelmien ratkaisemiseen, kuten onko sähköposti roskapostia vai ei.
- [Decision Trees](https://scikit-learn.org/stable/modules/tree.html): Päätöspuut ovat graafisia malleja, jotka esittävät joukkoa sääntöjä päätöksentekoon ja luokittelutehtäviin. Sitä käytetään usein luokitteluongelmien ratkaisemiseen.
- [Random Forests](https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestClassifier.html): Satunnaismetsät koostuvat useista päätöspuista, jotka ennustavat yhtä tulosarvoa. Sitä käytetään usein ennustamiseen ja luokitteluongelmien ratkaisemiseen.
- [LSTM-algoritmi](https://www.tensorflow.org/api_docs/python/tf/keras/layers/LSTM): Long Short-Term Memory (LSTM) on erityinen tyyppi neuroverkkoa, joka kuuluu RNN (Recurrent Neural Network) -perheeseen. Se käyttää ohjattua oppimista opetettaessaan ennustamaan tulevia arvoja aikasarjoissa.

### Seuraavaksi muodostetaan koneoppimisalgoritmi ennustamaan osakemarkkinoiden kehitystä

### Käytämme mallinnukseen LSTM nimistä neuroverkkoa
[LSTM](https://www.tensorflow.org/api_docs/python/tf/keras/layers/LSTM) (Long Short-Term Memory) on rekursiivinen syväoppimismalli, joka on erikoistunut sekvenssien, kuten aikasarjojen, käsittelyyn. Tämä tekee siitä hyödyllisen osakurssien tulevaisuuden ennustamiseen, koska osakurssit ovat aikasarjoja ja niiden tulevaisuus riippuu vahvasti historiallisista tiedoista.

LSTM:n kyky käsitellä pitkäaikaista muistia ja tunnistaa monimutkaisia riippuvuuksia aikasarjojen välillä on yksi sen suurimmista eduista ennustamisessa. Se kykenee havaitsemaan kausaalisia suhteita ja trendejä, joita lineaariset mallit ja satunnaismetsät eivät välttämättä huomaa. Lisäksi, koska LSTM on rekursiivinen malli, se voi käsitellä sisäkkäisiä rakenteita, kuten lauseita, musiikkia ja muita sekvenssejä, jotka voivat olla haastavia muille algoritmeille.

Lisäksi LSTM hyödyntää muistisolujen avulla tietoja, joita se on oppinut aiemmissa aikayksiköissä, mikä tekee siitä hyvin soveltuvan ennustamaan aikasarjoja, joissa on epävakaa, satunnainen tai kausaalinen luonne.

Kaiken kaikkiaan LSTM on siis parempi valinta osakurssien ennustamiseen, koska se kykenee käsittelemään monimutkaisia aikasarjoja ja havaitsemaan trendejä ja kausaalisia suhteita paremmin kuin lineaariset mallit tai satunnaismetsät.

#### Lue lisää aiheesta [täältä](https://towardsdatascience.com/an-introduction-to-long-short-term-memory-networks-lstm-27af36dde85d)

### 2.1 Harjoitusdatan esikäsittely
LSTM-mallin rakentamiseksi meidän on erotettava osakekurssitietomme koulutussarjaksi ja testisarjaksi. Lisäksi normalisoimme tietomme niin, että kaikki arvot ovat välillä 0-1.

Mallissa käytämme osakekurssien "close" hintoja eli päätöshintoja. Luodaan harjoitusdata niin, että harjoitusdatana käytämme 80% datajoukosta.

```python
# Tuodaan tarvittavat kirjastot
from sklearn.preprocessing import MinMaxScaler
import numpy as np
import math
```

```python
# Määritä koulutusdatan ja testidatan koko
train_size = int(0.8 * tesla_data.shape[0])

# Valitaan aikasarjan ensimmäiset 80% havaintoa koulutusdataksi ja loput testidataksi
training_set = tesla_data.iloc[:train_size, 1:2]

# Käytetään Scikit-Learn MinMaxScaleria normalisoidaksesi 
# kaikki varastotietomme välillä 0–1. 
# Muokataan myös normalisoidut tiedot kaksiulotteiseksi taulukoksi.
sc = MinMaxScaler(feature_range = (0, 1))
training_set_scaled = sc.fit_transform(training_set.values)

# Luo aikasarjadatan rakenteen, jossa käytetään 60 aikasaraketta ennustamaan seuraavaa arvoa
X_train = []
y_train = []
for i in range(60, train_size):
    X_train.append(training_set_scaled[i-60:i, 0]) # Otetaan 60 edeltävää arvoa
    y_train.append(training_set_scaled[i, 0]) # Ennustettava arvo

# Muunnetaan listat NumPy-taulukoiksi, jotta koneoppimisalgoritmi osaa käsitellä niitä
X_train, y_train = np.array(X_train), np.array(y_train)

# Muotoillaan vielä aikasarjadatan rakenne sopivaksi LSTM-mallin kouluttamista varten
# (tämä on Tensorflown hyväksymä muoto)
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
```

### 2.2 Testijoukon valmisteleminen

```python
# Ensimmäinen rivi valitsee tesla_data-datakehyksestä rivejä, jotka alkavat train_size-indeksistä
# ja ovat toisessa sarakkeessa.
test_set = tesla_data.iloc[train_size:, 1:2]

# dataset_total-datakehys luodaan yhdistämällä training_set-datakehys ja test_set-datakehys
dataset_total = pd.concat((training_set, test_set), axis = 0)

# Muuttuja inputs saa arvokseen dataset_total-datakehyksen viimeisten test_set-rivien sekä 60 edellisen rivin arvot
inputs = dataset_total[len(dataset_total) - len(test_set) - 60:].values

# Muotoillaan inputs-muuttuja yhden sarakkeen muotoiseksi
inputs = inputs.reshape(-1,1)

# Normalisoidaan inputs-muuttujan arvot sc-oliolla, joka on aiemmin luotu StandardScaler-olio
inputs = sc.transform(inputs)

# Alustetaan tyhjä lista X_test
X_test = []

test_size = tesla_data.shape[0] - train_size

# For-silmukka lisää inputs-muuttujan rivejä X_test-listaan.
for i in range(60, 60+test_size):
    # Lisätään inputs-listan i-60:i arvot listalle X_test
    X_test.append(inputs[i-60:i, 0])

# Muunnetaan lista NumPy-taulukoksi ja muotoillaan se vielä sopivaksi
X_test = np.array(X_test)
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))

# Tulostetaan X_test:n muoto
print(X_test.shape)
```

### 2.3 LSTM-verkkoarkkitehtuurin määrittäminen
Nyt olemme valmiita käyttämään avoimen lähdekoodin koneoppimiskirjastoa [Tensorflow](https://www.tensorflow.org/learn):ta LSTM-verkkoarkkitehtuurimme määrittämiseen.

Seuraava koodi luo syvän neuroverkon (deep neural network) regressiomallin (RNN) aikasarjojen ennustamiseen. Malli on toteutettu käyttäen useita LSTM-kerroksia (Long Short-Term Memory), joka mahdollistaa muistin ylläpitämisen aiemmista aikaväleistä ja auttaa ennustamaan tulevaisuuden arvoja.

Konkreettisesti koodi lisää neljä LSTM-kerrosta, joista jokainen sisältää 50 muistisolua (units). Jokainen kerros palauttaa sekvenssin (return_sequences = True), koska tarvitsemme kaikki kerrokset ennustamaan tulevaisuuden arvoja. Koodi luo myös output-kerroksen, joka on tiheä kerros (Dense), joka palauttaa yhden lopputuloksen (units = 1).

```python
# Tuodaan tarvittavat kirjastot
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
# Luodaan uusi Keras-Sequential-malli
model = keras.Sequential()

# Lisätään ensimmäinen LSTM-kerros
model.add(layers.LSTM(units = 50, return_sequences = True, input_shape = (X_train.shape[1], 1)))

# Lisätään toinen LSTM-kerros
model.add(layers.LSTM(units = 50, return_sequences = True))

# Lisätään kolmas LSTM-kerros
model.add(layers.LSTM(units = 50, return_sequences = True))

# Lisätään neljäs LSTM-kerros
model.add(layers.LSTM(units = 50))

# Lisätään lopullinen kerros, joka on tavallinen tiheäkerros (Dense), jolla on yksi lähtöyksikkö
model.add(layers.Dense(units = 1))

# Kootaan RNN-kerrokset yhteen ja määritetään menetelmä (optimizer) ja tappiofunktio (loss)
model.compile(optimizer = 'adam', loss = 'mean_squared_error')

# Sovitetaan RNN-malli koulutusdataan
model.fit(X_train, y_train, epochs = 100, batch_size = 32)
```

### 2.4 Harjoitetaan LSTM malliamme

#### Mallia laadittaessa annamme nämä kolme olennaista parametria:

- [optimizer](https://www.geeksforgeeks.org/intuition-of-adam-optimizer/) – Tämä on menetelmä, joka auttaa optimoimaan kustannusfunktion käyttämällä gradienttilaskua.
- [loss](https://www.geeksforgeeks.org/ml-common-loss-functions/) – Tappiofunktio, jolla seuraamme, kehittyykö malli harjoittelun myötä vai ei.
- [metrics](https://www.geeksforgeeks.org/metrics-for-machine-learning-model/) – Tämä auttaa arvioimaan mallia ennustamalla koulutusta ja validointitietoja.

```python
predicted_stock_price = model.predict(X_test)
predicted_stock_price = sc.inverse_transform(predicted_stock_price)
```

### Seuraavaksi arvoidaan mallin tarkkuutta

```python
# Käytetään MSE sekä MAE kaavoja
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

mse = mean_squared_error(test_set, predicted_stock_price)
mae = mean_absolute_error(test_set, predicted_stock_price)
print("MSE: ", mse)
print("MAE: ", mae)
```
